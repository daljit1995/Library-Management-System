```python
# Library Management System
from tkinter import *
from tkinter import messagebox
import mysql.connector

mydb=mysql.connector.connect(user='root',host='localhost')
mycursor=mydb.cursor()
mycursor.execute('use newbatch')
mycursor.execute('delete from lms where counts=0')
mydb.commit()

def add(): # for designing add data in sql
    def fxn():
        mycursor.execute('delete from lms where counts=0')
        mydb.commit()

        id=e1.get()
        auth=e2.get()
        book=e3.get()
        mycursor.execute('select author,book from lms')
        res=mycursor.fetchall()
        if (auth,book) in res:
            cmd=f"update lms set counts=counts+1 where author='{auth}' and book='{book}'"
        else:
            cmd=f"insert into lms values({id},'{auth}','{book}',1)"
        try:
            mycursor.execute(cmd)
            mydb.commit()
            messagebox.showinfo(title='Message',message='Success!')
        except:
            messagebox.showwarning(title='Message',message='Please try again')

        
    win=Toplevel()
    win.geometry('250x250')
    Label(win,text='Add Book',font=('Ariel',15,'bold'),fg='violet').place(x=10,y=10)
    Label(win,text='ID',font=('Ariel',10)).place(x=10,y=50)
    e1=Entry(win,width=20)
    e1.place(x=60,y=50)
    
    Label(win,text='Author',font=('Ariel',10)).place(x=10,y=100)
    e2=Entry(win,width=20)
    e2.place(x=60,y=100)

    Label(win,text='Book',font=('Ariel',10)).place(x=10,y=150)
    e3=Entry(win,width=20)
    e3.place(x=60,y=150)
    
    Button(win,text='Submit',command=fxn).place(x=20,y=200)
    win.configure(bg='black')

def dlt(): # for designing delete data in sql
    def fxn():
        mycursor.execute('delete from lms where counts=0')
        mydb.commit()

        id=e1.get()
        cmd=f'delete from lms where id={id}'
        try:
            mycursor.execute(cmd)
            mydb.commit()
            messagebox.showinfo(title='Message',message='Success!')
        except:
            messagebox.showwarning(title='Message',message='Please try again')
    
        
    win=Toplevel()
    win.geometry('250x250')
    
    Label(win,text='Delete Book',font=('Ariel',15,'bold'),fg='violet').place(x=10,y=10)
    Label(win,text='ID',font=('Ariel',10)).place(x=10,y=40)
    e1=Entry(win,width=20)
    e1.place(x=60,y=40)
    
    Button(win,text='Submit',command=fxn).place(x=20,y=80)
    win.configure(bg='black')
    
def issue():
    def fxn():
        mycursor.execute('delete from lms where counts=0')
        mydb.commit()

        name=e1.get()
        stuname=e2.get()
        UID=int(e3.get())
        auth=e4.get()
        mycursor.execute('select * from student')
        res=mycursor.fetchall()
        if (name,stuname,UID,auth) in res:
            messagebox.showwarning(message='Overlimt!')
        else:
            mycursor.execute('select book,author from lms')
            res2=mycursor.fetchall()
            if (name,auth) in res2:
                mycursor.execute(f"update lms set counts=counts-1 where author='{auth}' and book='{name}'")
                cmd=f"insert into student values('{name}','{stuname}',{UID},'{auth}')"
                try:
                    mycursor.execute(cmd)
                    mydb.commit()
                    messagebox.showinfo(title='Message',message='Success!')
                except:
                    messagebox.showwarning(title='Message',message='Please try again')
            else:
                messagebox.showwarning(title='Message',message='Book not Available')
                
                
       
        
    win=Toplevel()
    win.geometry('300x300+400+250')
    
    Label(win,text='Book Issue Detail',font=('Ariel',15,'bold'),fg='violet').place(x=10,y=10)
    Label(win,text='Book Name',font=('Ariel',10)).place(x=10,y=40)
    e1=Entry(win,width=20)
    e1.place(x=80,y=40)
    
    Label(win,text='Stu Name',font=('Ariel',10)).place(x=10,y=60)
    e2=Entry(win,width=20)
    e2.place(x=80,y=60)
    
    Label(win,text='U ID',font=('Ariel',10)).place(x=10,y=80)
    e3=Entry(win,width=20)
    e3.place(x=80,y=80)
    
    Label(win,text='Author',font=('Ariel',10)).place(x=10,y=100)
    e4=Entry(win,width=20)
    e4.place(x=80,y=100)
    
    Button(win,text='Submit',command=fxn).place(x=20,y=120)
    win.configure(bg='black')
         

root=Tk()
root.geometry('300x300+800+250')

Label(root,text='Welcome to library',bg='violet',font=('Ariel',20,'bold')).place(x=10,y=10)
var=IntVar()
r1=Radiobutton(root,text='Add',font=('Ariel',10,'bold'),variable=var,value=1,command=add)
r1.place(x=10,y=100)

r2=Radiobutton(root,text='Delete',font=('Ariel',10,'bold'),variable=var,value=2,command=dlt)
r2.place(x=10,y=130)

r3=Radiobutton(root,text='Issue',font=('Ariel',10,'bold'),variable=var,value=3,command=issue)
r3.place(x=10,y=160)


root.mainloop()
mycursor.close()
```
